```python
   P      =      V      *       I    =     I^2   *  R 
 (전력)        (전압)       (전류)

   V      =      I      *       R
 (전압)        (전류)       (저항)

 E = P * T =  I * T * V = Q * V(단위 : Wh)

 I   =  dq  /   dt     (전하 Q = I * T).
전류 = 전하 / 시간


전력 (P): 1초 동안에 소비하는 전기에너지. 단위 : W(와트). 전력은 보존

전압 (V): 전기 위치 에너지의 차이

전류 (I) (current): 같은 시간 동안 흘러가는 전기의 양. 단위 : A(암페어) 1A는 1초동안 1C의 전하가 이동.

(C는 전하 : 물체가 가지고 있는 전기적인 성질)

저항 (R) : 전기의 흐름이 윤활하지 않음을 나타낸 수치
```
