물리 계층 : 

데이터링크 계층 : peer to peer 간의 물리적 이동의 신뢰성(바로 옆) - error control(잘못 도착했으면 다시 보내도록), MAC(Multiple Access Control)(무선을 예로 들었을 때, 바로 옆에서 통신하는거 막기)

물리, 데이터링크 계층 : 내부 네트워크라고 칭함. (ethernet)

네트워크 계층{IP} : 게이트웨이에서 목적지까지 라우터들의 길을 선택해준다.

트랜스포트(전송) 계층{TCP/UDP} : 통신 양단간에 신뢰성 보장. 전체적인 길을 결정해놓는다고 보면 됨.(내 서버에서 아마존) 보낼 때랑 받을 때만 쓴다.
